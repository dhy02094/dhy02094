```python
# 

C = int(input())
for i in range(C):
    a = list(map(int,input().split()))
    b = a[0]
    c = a[1:]
    d = sum(c)/b
    e = []
    for i in c :
        if i > d:
            e.append(i)
    ans = len(e)/b * 100
    print(f'{ans:.3f}%')
    
    
        
        
```

```python
#   

n = int(input())

for _ in range(n):
    nums = list(map(int, input().split()))
    avg = sum(nums[1:])/nums[0]  #   (nums[0]: , nums[1:] )
    cnt = 0
    for score in nums[1:]:
        if score > avg:
            cnt += 1  #    
    rate = cnt/nums[0] *100
    print(f'{rate:.3f}%')
```
